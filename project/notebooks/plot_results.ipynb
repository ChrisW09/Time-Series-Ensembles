```python
import pandas as pd

def filter_ensemble_and_naive(df):
    # Define keywords for ensemble models and naive model
    ensemble_keywords = ['Ensemble', 'Meta Ensemble', 'Weighted Ensemble']
    naive_keyword = 'Naive'

    # Filter rows based on model names containing ensemble keywords or naive keyword,
    # but exclude 'Naive (drift)'
    filtered_df = df[df['Model'].str.contains('|'.join(ensemble_keywords)) | 
                     ((df['Model'].str.contains(naive_keyword)) & 
                      (~df['Model'].str.contains('Naive \(drift\)')))]

    return filtered_df

df_1 = filter_ensemble_and_naive(df1)
df_2 = filter_ensemble_and_naive(df2)
df_3 = filter_ensemble_and_naive(df3)
df_4 = filter_ensemble_and_naive(df4)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have four dataframes named df1, df2, df3, and df4

# Plotting MAE
plt.figure(figsize=(10, 6))
plt.plot(df_1['Model'], df_1['mape'], label='Dataset 1')
plt.plot(df_2['Model'], df_2['MAPE'], label='Dataset 2')
plt.plot(df_3['Model'], df_3['MAPE'], label='Dataset 3')
plt.plot(df_4['Model'], df_4['MAPE'], label='Dataset 4')
plt.xlabel('Model')
plt.ylabel('mae')
plt.xticks(rotation=90)
plt.title('Mean Absolute Error (MAE)')
plt.legend()
plt.tight_layout()
plt.show()
```
