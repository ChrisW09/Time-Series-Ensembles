```python
import pandas as pd
```

```python
# Read data
df = pd.read_csv('eurusd_df.csv')
df.info()
```

```python
display(df)
```

```python
# Count the number of rows with at least one NaN value
num_rows_with_nan = df.isnull().any(axis=1).sum()

# Display the result
print("Number of rows with at least one NaN value:", num_rows_with_nan)
print(f"Share of rows with at least one NaN values: {round(num_rows_with_nan/len(df),4)}")
```

```python
data = df.dropna()
```

```python
data
```

```python
# Validation for nan removal
if data.isnull().sum().sum() == 0:
    print("There are no NaN or null values in the DataFrame after cleaning.")
else:
    print("There are still NaN or null values in the DataFrame after cleaning.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Plot 'bid_open' as a line plot
plt.figure(figsize=(10, 6))
plt.plot(data.index, data['bid_open'], label='Bid Open', color='blue')
plt.title('Bid Open over Time')
plt.xlabel('Index')
plt.ylabel('Bid Open')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Plot all variables in one plot
plt.figure(figsize=(10, 6))
data.plot(ax=plt.gca())  # Passing the current axis to plot all columns
plt.title('Variables over Time')
plt.xlabel('Index')
plt.ylabel('Value')
plt.grid(True)
plt.legend(loc='upper left')  # You can adjust the location of the legend as needed
plt.show()
```
